### Act 10 Proyecto de simulación de inteligencia artificial  utiliza una lista de diccionarios,con if, else, elif  while tipo Menu

# Clase : Ingeniero

```python
# =====================================================================
# SISTEMA DE IA: PREDECTOR DE ÉXITO EN PROYECTOS DE INGENIERÍA
# =====================================================================

# Al inicio, la lista empieza vacía y el diccionario también
lista_ingenieros = []
losingenieros = {}

# Ciclo principal interactivo (Menú)
while True:
    print("\n" + "="*40)
    print("     🤖 CORE-AI: INGENIERÍA v1.0 🤖     ")
    print("="*40)
    print("1. Registrar datos de un Ingeniero")
    print("2. Ejecutar IA (Predicción de Éxito)")
    print("3. Mostrar base de datos actual")
    print("4. Salir del sistema")
    print("-"*40)

    opcion = input("Selecciona una opción (1-4): ")

    # OPCIÓN 1: Registrar datos en el diccionario y guardarlos en la lista
    if opcion == "1":
        print("\n--- REGISTRO DE DATOS ---")
        nombre = input("Nombre del ingeniero: ")
        especialidad = input("Especialidad (Sistemas / Civil / Mecánica): ")
        experiencia = int(input("Años de experiencia: "))
        proyecto = input("Nombre del proyecto asignado: ")

        # Intercambio de xxxx por 'losingenieros' (Diccionario)
        losingenieros = {
            "nombre": nombre,
            "especialidad": especialidad,
            "experiencia": experiencia,
            "proyecto": proyecto
        }

        # Agregamos el diccionario a la lista de ingenieros
        lista_ingenieros.append(losingenieros)
        print(f"\n✅ Datos de {nombre} guardados exitosamente en la memoria de la IA.")

    # OPCIÓN 2: Procesamiento de la Inteligencia Artificial (Sistema Experto)
    elif opcion == "2":
        print("\n--- PROCESANDO ALGORITMO DE IA ---")

        # Validamos si la lista está vacía
        if len(lista_ingenieros) == 0:
            print("⚠️ Error de procesamiento: La base de datos está vacía. Registra un ingeniero primero.")
        else:
            print("Analizando variables, perfiles y entornos de riesgo...\n")

            # Ciclo FOR para recorrer la lista de diccionarios
            for ingeniero in lista_ingenieros:
                print(f"🔮 Evaluando a: {ingeniero['nombre']} | Proyecto: {ingeniero['proyecto']}")

                # Reglas de la IA usando IF, ELIF y ELSE
                if ingeniero['especialidad'].lower() == "sistemas" and ingeniero['experiencia'] > 5:
                    print("➡️ DIAGNÓSTICO IA: Éxito del 95%. Alta probabilidad de automatización eficiente.")

                elif ingeniero['especialidad'].lower() == "civil" and ingeniero['experiencia'] < 3:
                    print("➡️ DIAGNÓSTICO IA: Riesgo Moderado. Se sugiere supervisión en el cálculo estructural.")

                elif ingeniero['especialidad'].lower() == "mecánica":
                    print("➡️ DIAGNÓSTICO IA: Éxito del 85%. Optimización de maquinaria y termodinámica estables.")

                else:
                    print("➡️ DIAGNÓSTICO IA: Perfil estándar. Éxito estimado en un 70%. Resultados dentro del promedio.")
                print("-" * 50)

    # OPCIÓN 3: Mostrar los datos almacenados
    elif opcion == "3":
        print("\n--- BASE DE DATOS DE INGENIEROS ---")
        if len(lista_ingenieros) == 0:
            print("La lista está vacía.")
        else:
            # Ciclo FOR para mostrar de forma limpia cada diccionario de la lista
            for ingeniero in lista_ingenieros:
                print(f"👷 {ingeniero['nombre']} | Rama: {ingeniero['especialidad']} | Experiencia: {ingeniero['experiencia']} años | Proyecto: {ingeniero['proyecto']}")

    # OPCIÓN 4: Romper el ciclo While y salir
    elif opcion == "4":
        print("\nApagando los servidores de IA... ¡Hasta la próxima!")
        break

    # Filtro por si el usuario presiona otra tecla
    else:
        print("❌ Opción no válida. Por favor, selecciona un número del 1 al 4.")
print("Programa realizado por David Barrera 1388")
```
